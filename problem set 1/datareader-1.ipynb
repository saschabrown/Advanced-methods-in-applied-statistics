```python
# Get the data from the website
url = "https://www.nbi.dk/~koskinen/Teaching/AdvancedMethodsInAppliedStatistics2018/data/authors-acknowledgements-v5.pdf"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

# Find the table containing the data
table = soup.find('table')
table
```

```python
# in the code bellow we are trying to get the headers of the table, the headers are the second row of the table
headers = []
for header in table.find_all('tr')[1].find_all('th'):
    headers.append(header.text.strip())


headers = headers[1:] # remove the first header, which is empty
```

```python
with open('basketball_data.csv', 'w', newline='') as csvfile:
    # Create a CSV writer object
    writer = csv.writer(csvfile)
    
    # Extract data from table rows
    for row in table.find_all('tr'):
        # Extract data from table cells (td)
        columns = row.find_all('td')
        # Write the data to the CSV file
        writer.writerow([column.text.strip() for column in columns])
df = pd.read_csv('basketball_data.csv', header = None)

start_column_index = 1
end_column_index = 21


df = df.iloc[:, start_column_index:end_column_index]


column_to_string = 0  # Adjust this according to your requirement
df.iloc[:,column_to_string] = df.iloc[:,column_to_string].astype(str)
#Now i remove all the numbers in the team name
df.iloc[:,0] = [re.sub(r' \d+$', '', item) for item in df.iloc[:,0]]
```

```python
types =np.array(df.dtypes)
new_headers = []
first_row =  df.iloc[1]

count = 0

for i in range(len(types)):
    
    if types[i] == 'int64':
        count += 1
        new_headers.append(str(headers[i-count])+ ' rank')
    else:
        new_headers.append(headers[i-count])


```

```python

name_counts = {}

# Iterate through the new headers
for i, name in enumerate(new_headers):
    # If the name has already appeared
    if name in name_counts:
        # Increment the count
        name_counts[name] += 1
        # If it's the second appearance, add 'strengh' to the name
        if name_counts[name] == 2:
            new_headers[i] += ' strength'
    else:
        # If it's the first appearance, initialize the count
        name_counts[name] = 1


#I'm now giving new names to my columns as i found them
    
df.columns = new_headers

# i Know save this as a csv file

df.to_csv('updated_dataframe_2009', index=False)
```
